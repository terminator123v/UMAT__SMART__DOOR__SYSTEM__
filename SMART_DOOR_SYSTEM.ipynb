```python
class Staff:
    def __init__(self, s_name: str, access_code: str):
        self.s_name = s_name
        self.access_code = access_code

    @property
    def access_code(self) -> str:
        return self.__access_code

    @access_code.setter
    def access_code(self, new_code: str):
        if not isinstance(new_code, str):
            raise TypeError("Access code must be a string.")
        if len(new_code) < 6:
            raise ValueError("Validation Error: Access code must be at least 6 characters long.")
        self.__access_code = new_code

    def view_access_code(self, requester_role: str) -> str:
        authorized_roles = ["admin", "dean", "dean of oric", "manager"]
        if requester_role.lower() in authorized_roles:
            return self.access_code
        else:
            return "Access Denied: You are not authorized to view this access code."

    def display_staff_info(self):
        masked_code = "*" * (len(self.__access_code) - 2) + self.__access_code[-2:] if len(self.__access_code) > 2 else "******"
        print(f"\n--- Staff Details ---")
        print(f"Staff Name: {self.s_name}")
        print(f"Access Code Status: {masked_code} (Protected)")
        print("---------------------")


if __name__ == "__main__":
    print("--- Initializing Staff Member ---")
    try:
        staff = Staff("Dr. Botwey Bright", "ORIC2026")
        staff.display_staff_info()

        print("\n[Attempt 1] Student attempts to view the access code:")
        print(staff.view_access_code("Student"))

        print("\n[Attempt 2] Dean of ORIC attempts to view the access code:")
        print(f"Retrieved Code: {staff.view_access_code('Dean of ORIC')}")

        print("\n[Attempt 3] Attempting to update access code to '123' (Invalid):")
        staff.access_code = "123"

    except ValueError as e:
        print(f"Caught expected error: {e}")

    try:
        print("\n[Attempt 4] Attempting to update access code to 'UMaT_9876' (Valid):")
        staff.access_code = "UMaT_9876"
        staff.display_staff_info()
        print(f"Updated Code: {staff.view_access_code('Admin')}")

    except Exception as e:
        print(f"Unexpected error: {e}")
```
